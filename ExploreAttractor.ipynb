```python
from attractor import Attractor
%matplotlib inline
```

Loads the Attractor class from the attractor file

```python
a = Attractor()
a.evolve()
a.plot3d()
a.evolve(order=2)
a.plot3d()
a.evolve(order=1)
a.plot3d()
```

Plots the 3d plot of attractor with the initial values for each order of Runge Kutta. It appears the the higher the order, the "fuller" the graph looks. The fourth order graph (1st one) has the most detail and looks as if it makes features of the 1st order and 2nd order graphs more prominent. For example in the 1st order graph (3rd) there is a hallow circle in the middle and a singular line on the outside, and in the 2nd order graph (2nd) there is a more prominent distinction between the outer and inner circles but there isn't much of a hallow circle. The 4th order graph (1st) appears to combine these features and has a distinguished inner and outer circles along with a hallow circle in the center.

```python
a.evolve()
a.plotx()
a.ploty()
a.plotz()
a.evolve(order=2)
a.plotx()
a.ploty()
a.plotz()
a.evolve(order=1)
a.plotx()
a.ploty()
a.plotz()
```

Plots x(t), y(t), z(t) with initial conditions for each Runge Kutta order.

X and Y seem to behave almost identically (within each order), while Z is more symmetric.

There doesn't seem to be much of a difference between the orders except for when the graphs converge.

```python
a.evolve()
a.plotxy()
a.plotyz()
a.plotzx()
a.evolve(order=2)
a.plotxy()
a.plotyz()
a.plotzx()
a.evolve(order=1)
a.plotxy()
a.plotyz()
a.plotzx()
```

Plots x-y, y-z, z-x planar projections.

Each of these plots looks like a specific view of the 3D plot and highlights the differences between the orders I noticed in the 3D plot.

### Now let's change the initial values and see how the plots differ:

```python
a.evolve([10,20,30])
a.plot3d()
```

Increasing the initial conditions appear to make the outside lines more defined with a greater amount of space between them present.

```python
a.evolve([10,20,30])
a.plotx()
a.ploty()
a.plotz()
```

```python
a.evolve([10,20,30])
a.plotxy()
a.plotyz()
a.plotzx()
```

```python
a.evolve([100,100,100])
a.plot3d()
```

If you increase the initial conditions high enough the 'bowtie' turns into a line in 3 dimensional space.

```python
a.evolve([0.01,0.0001,25])
a.plot3d()
```

```python
a.evolve([0.01,0.0001,25], 1)
a.plot3d()
```

```python
a.evolve([25,25,25])
a.plot3d()
```

```python
a.evolve([100,100,100])
a.plotx()
a.ploty()
a.plotz()
```

```python
a.evolve([100,100,100])
a.plotxy()
a.plotyz()
a.plotzx()
```

If you increase the initial conditions high enough all of the graphs become a line.

```python
a.evolve([10,20,40])
a.plot3d()
```

```python
a.evolve([10,20,40])
a.plotxy()
a.plotyz()
a.plotzx()
```

The outer circles seem to appear to converge back towards the middle circles
